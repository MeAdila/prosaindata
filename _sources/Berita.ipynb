```python
import pandas as pd
import requests as re
from bs4 import BeautifulSoup as bs
from urllib.request import Request, urlopen
import csv
```

```python
header = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
url = 'https://www.cnnindonesia.com/search/?query=ridwan+kamil'

judul = []
konten = []

for page in range(1,9):
  req = re.get(url+str(page), headers=header)
  soup = bs(req.text, 'html.parser')
  items = soup.findAll('article')
  for it in items:
    try : link = it.find('a')['href']
    except : link =''
    try : title = it.find('h2', 'title').text
    except : title =''
    judul.append(title)

    if link != '':
      req2 = re.get(str(link), headers=header)
      soup2 = bs(req2.text, 'html.parser')
      items2 = soup2.findAll("div", "detail_text")
      for it2 in items2:
        isi = it2.find('p').text

        if isi != '':
          konten.append(isi)


JudulBerita = pd.DataFrame(judul, columns=["Judul Berita"])
IsiBerita = pd.DataFrame(konten, columns=["Isi Berita"])

data = pd.concat([JudulBerita, IsiBerita], axis=1)
data = data.loc[0:50]
data
```
