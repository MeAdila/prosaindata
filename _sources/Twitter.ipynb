```python
import twint
import nest_asyncio
nest_asyncio.apply()
c = twint.Config()
c.Search = "ridwankamil"
# c.Since = "2021-10-11"
# c.Until = "2021-10-12"
# c.Username = "#ridwankamil"
c.Limit = 500
c.Store_csv = True
c.Custom["tweet"] = ["tweet"]
c.Output = "Data_ridwan.csv"
twint.run.Search(c)
```
